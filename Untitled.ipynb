```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
dataset = pd.read_csv('water_potability.csv')
dataset.shape
```

```python
dataset.head(5)
```

```python
dataset.ph = dataset['ph'].replace(np.nan, dataset.ph.mean()).astype(float)
dataset.Hardness = dataset['Hardness'].replace(np.nan, dataset.Hardness.mean()).astype(float)
dataset.Solids = dataset['Solids'].replace(np.nan, dataset.Solids.mean()).astype(float)
dataset.Chloramines = dataset['Chloramines'].replace(np.nan, dataset.Chloramines.mean()).astype(float)
dataset.Sulfate = dataset['Sulfate'].replace(np.nan, dataset.Sulfate.mean()).astype(float)
dataset.Conductivity = dataset['Conductivity'].replace(np.nan, dataset.Conductivity.mean()).astype(float)
dataset.Organic_carbon = dataset['Organic_carbon'].replace(np.nan, dataset.Organic_carbon.mean()).astype(float)
dataset.Trihalomethanes = dataset['Trihalomethanes'].replace(np.nan, dataset.Trihalomethanes.mean()).astype(float)
dataset.Turbidity = dataset['Turbidity'].replace(np.nan, dataset.Turbidity.mean()).astype(float)
```

```python
X = dataset.iloc[:,[0,8]].values
y = dataset.iloc[:,9].values


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= .3, random_state = 0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

lista = ('rbf', 'linear', 'poly', 'sigmoid')
for x in lista:
    classifier = SVC(kernel = x , random_state=0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    from sklearn.metrics import confusion_matrix, accuracy_score
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    print(accuracy_score(y_test,y_pred))
    
    from matplotlib.colors import ListedColormap
    X_set, y_set = X_test, y_test
    X1,X2 = np.meshgrid(np.arange(start = X_set[:,0].min() - 1, stop = X_set[:,0].max() +1, step = 0.01),
                       np.arange(start = X_set[:,1].min() - 1, stop = X_set[:,1].max() +1, step = 0.01))
    plt.contourf(X1,X2, classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha = 0.75,
                cmap = ListedColormap(('black','red','blue')))
    plt.xlim(X1.min(),X1.max())
    plt.ylim(X2.min(),X2.max())
    for i,j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set==j,1],
                   c = ListedColormap(('black','red','blue'))(i), label= j)

    plt.title('SVM (Test set)')
    plt.xlabel('Age')
    plt.ylabel('sdfs')
    plt.legend()
    plt.show()
```

```python
target_names = ['nonpotable','potable']
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=1)
X_r2 = lda.fit(X, y).transform(X)

print('explained variance ratio (first two components in PCA):', str(pca.explained_variance_ratio_))
```

```python
lw = 2
colors = ['red', 'blue']

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of this dataset')
plt.figure()
```
